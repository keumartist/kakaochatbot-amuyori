### json recipe 파일을 csv recipe 파일로 변환하기

```python
import pandas as pd
import json
```

```python
data = {} 
with open('./recipe.json') as json_file:
    data = json.load(json_file)
#     print(data)
```

```python
data['COOKRCP01']['row']
```

```python
dataframe = pd.DataFrame([data['COOKRCP01']['row'][0]])
for recipe in data['COOKRCP01']['row']:
    d = pd.DataFrame([recipe])
    dataframe = pd.merge(dataframe, d, how="outer")
```

```python
data2 = {} 
with open('./recipe2.json') as json_file:
    data2 = json.load(json_file)

dataframe2 = pd.DataFrame([data2['COOKRCP01']['row'][0]])
for recipe in data2['COOKRCP01']['row']:
    d = pd.DataFrame([recipe])
    dataframe2 = pd.merge(dataframe2, d, how="outer")
```

```python
data3 = {} 
with open('./recipe3.json') as json_file:
    data3 = json.load(json_file)

dataframe3 = pd.DataFrame([data3['COOKRCP01']['row'][0]])
for recipe in data3['COOKRCP01']['row']:
    d = pd.DataFrame([recipe])
    dataframe3 = pd.merge(dataframe3, d, how="outer")
```

```python
recipe_df = pd.merge(dataframe, dataframe2, how="outer")
```

```python
recipe_df = pd.merge(recipe_df, dataframe3, how="outer")
```

```python
# 이미지 경로 이상한 애들 확인
gogo = recipe_df['ATT_FILE_NO_MAIN'][847]
gogo
```

```python
sorted_recipe_df = recipe_df.reindex(sorted(recipe_df.columns), axis=1)
```

```python
sorted_recipe_df.to_csv("./recipe.csv")
```
